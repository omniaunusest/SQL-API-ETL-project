```python

# Conexión a la base de datos
db = mysql.connector.connect(
    host="localhost",
    user="tu_usuario",
    password="tu_contraseña",
    database="MusicStreamStupendo"
)
cursor = db.cursor()

# Crear tablas
cursor.execute("""
CREATE TABLE IF NOT EXISTS artistas (
    id_artista VARCHAR(45) PRIMARY KEY,
    nombre VARCHAR(100),
    biografia TEXT,
    playcount INT,
    listeners INT,
    artistas_similares JSON
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS generos (
    id_genero INT AUTO_INCREMENT PRIMARY KEY,
    nombre VARCHAR(50)
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS albums (
    id_album VARCHAR(50) PRIMARY KEY,
    nombre VARCHAR(100),
    año_lanzamiento DATE,
    popularidad INT,
    id_artista VARCHAR(45),
    FOREIGN KEY (id_artista) REFERENCES artistas(id_artista)
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS canciones (
    id_cancion INT PRIMARY KEY,
    titulo VARCHAR(100),
    año_lanzamiento VARCHAR(45),
    id_genero INT,
    id_artista VARCHAR(45),
    FOREIGN KEY (id_genero) REFERENCES generos(id_genero),
    FOREIGN KEY (id_artista) REFERENCES artistas(id_artista)
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS artistas_generos (
    id_artista VARCHAR(45),
    id_genero INT,
    PRIMARY KEY (id_artista, id_genero),
    FOREIGN KEY (id_artista) REFERENCES artistas(id_artista),
    FOREIGN KEY (id_genero) REFERENCES generos(id_genero)
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS albums_canciones (
    id_album VARCHAR(50),
    id_cancion INT,
    PRIMARY KEY (id_album, id_cancion),
    FOREIGN KEY (id_album) REFERENCES albums(id_album),
    FOREIGN KEY (id_cancion) REFERENCES canciones(id_cancion)
)
""")

db.commit()
cursor.close()
db.close()
```
