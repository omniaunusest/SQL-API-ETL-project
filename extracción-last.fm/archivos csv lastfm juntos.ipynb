```python
import pandas as pd
import os

archivos_csv = ['country_2018_2022_canciones.csv', 'metal_2018_2022_canciones.csv', 'pop_2018_2022_canciones.csv','rb_2018_2022_canciones.csv', 'rock_2018_2022_canciones.csv']

ruta_carpeta = 'C:/Users/maria/Desktop/adalab/da-promo-60-modulo2-equipo3/csv'
ruta_completa = [os.path.join(ruta_carpeta, f) for f in archivos_csv]



lista_df = []
for archivo in ruta_completa:
    try:
        df = pd.read_csv(archivo)
        lista_df.append(df)
        print('archivo cargado')
    except:
        print('error')
    
df_combinado = pd.concat(lista_df, ignore_index= True)

print('todo ok')
print({df_combinado.shape})

archivo_final = 'artistas_lastfm_combinados_final.csv'
df_combinado.to_csv(os.path.join(ruta_carpeta, archivo_final), index=False)
print('archivo final guardado')
```
