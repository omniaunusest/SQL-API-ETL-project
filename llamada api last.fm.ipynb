```python
apikey = "b4fcada0293085a139072e5e87a365b4"   #aqui iria la apikey individual de cada una. No hace falta shared secret como en spotipy
url = "http://ws.audioscrobbler.com/2.0/"

# El artista que quieres analizar
artista_a_buscar = "slipknot"
```

```python
# --- PAR√ÅMETROS PARA artist.getInfo ---
params_info = {
    "method": "artist.getInfo",
    "artist": artista_a_buscar,
    "api_key": apikey,
    "format": "json",
    "autocorrect": 1 # Importante: corrige posibles errores tipogr√°ficos en el nombre del artista
}

# Realizar la llamada a la API
response_info = requests.get(url, params=params_info)
datos_info = response_info.json()

if response_info.status_code == 200 and 'artist' in datos_info:
    info_artista = datos_info['artist']
    
    # üìù Biograf√≠a y Tags (G√©neros)
    print("--- BIOGRAF√çA Y TAGS ---")
    print(f"Nombre: {info_artista.get('name')}")
    # La biograf√≠a viene con HTML y es extensa, a menudo querr√°s un resumen:
    biografia_corta = info_artista.get('bio', {}).get('summary', 'No disponible')
    print(f"Biograf√≠a (Resumen): {biografia_corta.split('<a href')[0]}") #el <a href ahora mismo no es importante pero mas adelante lo necesitaremos para el analisis de datos
    
    # üìà Estad√≠sticas de Reproducci√≥n y Popularidad
    print("\n--- ESTAD√çSTICAS ---")
    stats = info_artista.get('stats', {})
    print(f"Reproducciones (Playcount Total en Last.fm): {stats.get('playcount')}")
    print(f"N√∫mero de Oyentes √önicos: {stats.get('listeners')}")
else:
    print(f"‚ùå Error al obtener informaci√≥n del artista: {artista_a_buscar}")
    # print(datos_info)
#3. Obtener Artistas Similares (artist.getSimilar)
#Este m√©todo es ideal para encontrar artistas relacionados, lo cual es √∫til para la recomendaci√≥n o segmentaci√≥n en tu an√°lisis.

##Python

# --- PAR√ÅMETROS PARA artist.getSimilar ---
params_similar = {
    "method": "artist.getSimilar",
    "artist": artista_a_buscar,
    "api_key": apikey,
    "format": "json",
    "limit": 5 # Obtener los 5 artistas m√°s similares
}

# Realizar la llamada a la API
response_similar = requests.get(url, params=params_similar)
datos_similar = response_similar.json()

if response_similar.status_code == 200 and 'similarartists' in datos_similar:
    artistas_similares = datos_similar['similarartists']['artist']
    
    # ü§ù Artistas Similares
    print("\n--- ARTISTAS SIMILARES ---")
    if artistas_similares:
        for i, similar in enumerate(artistas_similares):
            # 'match' indica la similitud (de 0 a 1)
            print(f"{i+1}. {similar.get('name')} (Similitud: {similar.get('match')})")
    else:
        print("No se encontraron artistas similares.")
else:
    print(f"‚ùå Error al obtener artistas similares para: {artista_a_buscar}")
```
