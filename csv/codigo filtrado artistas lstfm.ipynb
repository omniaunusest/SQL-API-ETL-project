```python
import pandas as pd #simplificar y arreglar
import os


archivo_entrada = 'total_datos_artistas_lastfm_2018-2022.csv'
ruta_carpeta = 'C:/Users/maria/Desktop/adalab/da-promo-60-modulo2-equipo3/csv/generos'

ruta_archivo_completa = os.path.join(ruta_carpeta, archivo_entrada)
columna_artista = 'artista'# ¬°IMPORTANTE! Aseg√∫rate de que este sea el nombre real de tu columna de artistas

try:
    # 2. Cargar el archivo CSV en un DataFrame de Pandas
    df = pd.read_csv(ruta_archivo_completa)
    print(f"‚úÖ Archivo '{archivo_entrada}' cargado. Filas totales: {len(df)}")
    
    # 3. Eliminar duplicados bas√°ndose √∫nicamente en la columna del artista
    # 'keep="first"' asegura que se mantiene la primera aparici√≥n de cada artista.
    df_artistas_unicos = df.drop_duplicates(subset=[columna_artista], keep='first')
    
    # 4. Mostrar informaci√≥n y guardar el resultado
    print(f"‚ú® Proceso completado. Artistas √∫nicos encontrados: {len(df_artistas_unicos)}")
    
    archivo_salida = 'artistas_unicos_filtrados.csv'
    ruta_archivo_salida = os.path.join(ruta_carpeta, archivo_salida)
    
    # Guardar el nuevo DataFrame con solo artistas √∫nicos
    df_artistas_unicos.to_csv(ruta_archivo_salida, index=False)
    
    print(f"üíæ Archivo con artistas √∫nicos guardado en: {ruta_archivo_salida}")

except FileNotFoundError:
    print(f"‚ùå ERROR: El archivo '{archivo_entrada}' no se encontr√≥ en la ruta especificada.")
except Exception as e:
    print(f"‚ùå ERROR al procesar el archivo: {e}")
```
