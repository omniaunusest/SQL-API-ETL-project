```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd

auth_manager = SpotifyClientCredentials(client_id="4fe8c3ad5e9c4851a77b6fbe2912fa5c" , client_secret="62bbeef6c90d4de783fa926d0cd1b5a5")
sp = spotipy.Spotify(auth_manager=auth_manager)

años = range(2018, 2023)  # de 2018 a 2022
genero = "pop"
artistas = []
todos_los_tracks = []

print("🎶🎵Tracks de artistas del Pop de los años 2018-2022🎶🎵")

for año in años:
    
    query = f"genre:{genero}, year:{año}" #variable de la q porque sino sólo itera por un añ concreto
    for o in range(0,150, 50):
        print(f"Tracks del {año}", "-", o)
    try:
        resultados = sp.search(q=query, type="track", limit=50, offset=o)

        
        for track in resultados['tracks']['items']:
             nombre_artista = track['artists'][0]['name']
             track_data = {"✨Título de la Canción": track['name'],
                "🎭Artista": nombre_artista,
                "🎼Nombre del Álbum": track['album']['name'],
                "++ID del Álbum": track['album']['id'],
                "**Año de Publicación": track['album']['release_date'][:4]}
             artistas.append(nombre_artista)
             todos_los_tracks.append(track_data)
            #hasta aquí es más o menos las variables que hemos hecho 
            
    except:
        print(f"Error al buscar en {año}")

print(" 🔉🔊Total de artistas extraídos:🎙🧨", len(artistas))

#la tabla 
df_tracks = pd.DataFrame(todos_los_tracks) #previo me he hecho una variable vacia arriba 
print("--- 🔈🔉🔊Tracks Encontrados ---")
print(df_tracks.head())

# conteo de los arts y sus apariciones en listas 
conteo_artistas = {}  # queda  por desarrollarxxxx 
```
