```python
import json

# Cargar el JSON
with open('rock-extraccion-pruebas-json.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Generar el script SQL
with open('spotify_data.sql', 'w', encoding='utf-8') as sql_file:
    for record in data:
        columns = ', '.join(record.keys())
        values = ', '.join([f"'{str(value).replace("'", "''")}'" for value in record.values()])
        sql = f"INSERT INTO spotify_data ({columns}) VALUES ({values});\n"
        sql_file.write(sql)

print("Script SQL listo de una p*** vez.")
```
