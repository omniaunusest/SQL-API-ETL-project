```python
import pandas as pd
from sqlalchemy import create_engine

# --- 1. Definir los parámetros de tu Base de Datos MySQL ---
# ¡Asegúrate de cambiar estos datos por los tuyos!
usuario = 'root'
contraseña = 'AlumnaAdalab'
host = 'localhost' # O la IP de tu servidor
bbdd = 'musicstreamstupendo'
nombre_tabla_sql = 'artistas_last_fm_combinados' 

# --- 2. Crear el Motor (El Traductor) ---
# Esto genera una cadena de conexión estándar para SQLAlchemy
cadena_conexion = f"mysql+mysqlconnector://{usuario}:{contraseña}@{host}/{bbdd}"
engine = create_engine(cadena_conexion)

# --- 3. Ejecutar .to_sql() usando el Motor ---
# Asumo que tu DataFrame se llama 'artistas_last_fm_combinados'
try:
    df_artistas_last_fm_combinados.to_sql(
        nombre_tabla_sql,
        engine, # <-- ¡Usamos el motor de SQLAlchemy aquí!
        if_exists='replace',
        index=False 
    )

    print(f"\n✅ ¡Datos volcados con éxito en la tabla SQL '{nombre_tabla_sql}' usando SQLAlchemy!")

except Exception as e:
    print(f"Ocurrió un error al volcar los datos: {e}")
```
