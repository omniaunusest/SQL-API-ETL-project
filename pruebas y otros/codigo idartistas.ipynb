```python
import pandas as pd
import os

ruta_tabla1 = 'total_canciones_2018-2022.csv'
ruta_tabla2 = 'total_datos_artistas_lastfm_2018-2022.csv'
ruta_salida = 'artistas_con_id.csv'


df_tracks = pd.read_csv(ruta_tabla1, encoding='utf-8')
print(f"Tabla1 canciones Filas: {df_tracks.shape[0]}")

df_artistas = pd.read_csv(ruta_tabla2, encoding='utf-8')
print(f"Tabla2 artistas) cargada. Filas: {df_artistas.shape[0]}")


# Crear el DataFrame de Mapeo (ID + Artista)
# Seleccionamos las columnas necesarias: la clave de uniÃ³n ('artista') y el valor a aÃ±adir ('id').
df_mapeo = df_tracks[['id', 'artista']].copy()

df_mapeo = df_mapeo.rename(columns={'id': 'artista_id'})

#eliminamos duplicados
df_mapeo = df_mapeo.drop_duplicates(subset=['artista'])

print(f" Artistas Ãºnicos: {df_mapeo.shape[0]}")

# Realizamos la uniÃ³n
df_nuevo_artistas = pd.merge(
    df_artistas, 
    df_mapeo, 
    left_on='artista', 
    right_on='artista', # Se unen por la columna 'artista' que es comÃºn
    how='left' 
)

# Definimos el orden final de las columnas, moviendo el 'artista_id' al final
columnas_finales = [
    'artista', 
    'playcount',
    'listeners',
    'biografia_resumen',
    'similares',
    'artista_id' 
]

# Reordenamos el DataFrame
df_final = df_nuevo_artistas[columnas_finales].copy()


print(df_final.head())

df_final.to_csv(ruta_salida, index=False, encoding='utf-8')

print(f"\nðŸŽ‰ Â¡Ã‰xito! El nuevo CSV '{ruta_salida}' ha sido creado.")
```
