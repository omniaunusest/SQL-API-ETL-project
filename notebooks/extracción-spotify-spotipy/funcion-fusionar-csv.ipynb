```python
import random 
import pandas as pd
import os

# FUNCIÓN PARA FUSIONAR .CSV con rutas relativas

def fusionar_csv(ruta_origen, archivos, nombre_salida, ruta_salida):
   
    rutas = [os.path.join(ruta_origen, archivo) for archivo in archivos]
    lista_cosos = ['Charmander,', 'Pikachu,', 'Mamagüevo,', 'A ver, elegante,', 'Mi sentido pésame:', 'Error ZX:', 'Error P-206:', 'Error 4x4']
    Error = random.choice(lista_cosos)

    # ¿Los archivos de referencia existen?
    for ruta in rutas:
        if not os.path.exists(ruta):
            raise FileNotFoundError(f"El archivo {ruta} no existe. Verifica la ruta y el nombre.")

    # Fusionar
    try:
        dfs = [pd.read_csv(ruta) for ruta in rutas]
        df_unit = pd.concat(dfs, ignore_index=True)
    except Exception as e:
        raise RuntimeError(f"Error al leer o combinar los archivos: {e}")

    # Construimos la ruta completa del archivo de salida
    ruta_final = os.path.join(ruta_salida, nombre_salida)

    # ¿El archivo de salida ya existe?
    if os.path.exists(ruta_final):
        print((f"{Error} {nombre_salida} ya existe en {ruta_salida}."))
    else: 
        df_unit.to_csv(ruta_final, index=False)
        print(f"Tus archivos evolucionaron a {nombre_salida} en {ruta_final}.")
```

```python
# USO
ruta_origen = "../csv"
archivos = [
    "country_2018_2022_canciones.csv",
    "metal_2018_2022_canciones.csv",
    "rb_2018_2022_canciones.csv",
    "pop_2018_2022_canciones.csv",
    "rock_2018_2022_canciones.csv"
]
nombre_salida = "total_canciones_2018-2022.csv"
ruta_salida = "../csv"

fusionar_csv(ruta_origen, archivos, nombre_salida, ruta_salida)
```
